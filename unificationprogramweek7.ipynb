```python
def check_syntax(temp):
    incr = 0
    br = 0
    for i in temp:
        if i == "(":
            br += 1
        elif i == ")":
            br -= 1
        if i.isupper() and incr == 0:
            return "error"
        if i != ")" and incr == len(temp):
            return "error"
        incr += 1
    if br != 0:
        return "error"
    else:
        return "passed"

def check_const_var(temp):
    for i in temp:
        if i.islower():
            return "constant"
        else:
            return "variable"

def check_type(temp):
    if temp.isalpha() or "(" in temp or ")" in temp or "," in temp:
        if "(" in temp:
            if ")" in temp:
                if check_syntax(temp) == "error":
                    return "input error"
                else:
                    return "function"
            else:
                return "input error"
        if "(" not in temp and ")" not in temp and "," not in temp:
            return check_const_var(temp)
    else:
        return "input error"

def check_numarg(temparg):
    br = 0
    co = 0
    for i in temparg:
        if i == "(":
            br += 1
        elif i == ")":
            br -= 1
        if i == "," and br == 1:
            co += 1
    if br == 0:
        return co + 1
    else:
        return "input error"

def check_funname(temparg):
    funname = ""
    for i in temparg:
        if i == "(":
            break
        funname += i
    return funname

def ret_arg(temparg):
    arguments = []
    actual = ""
    br = 0
    co = 0
    inc = ""
    first = 0
    for j in temparg:
        if first == 1:
            actual += j
        if j == "(":
            first = 1
    actual = actual[:-1]
    for i in actual:
        if i == "," and br == 0:
            co += 1
            arguments.append(inc)
            inc = ""
        else:
            inc += i
        if i == "(":
            br += 1
        elif i == ")":
            br -= 1
    if co == 0:
        arguments.append(actual)
    else:
        arguments.append(inc)
    return arguments

def subst(v, l):
    for i in range(len(v)):
        if l[0][0] in v[i][1]:
            v[i][1] = v[i][1].replace(l[0][0], l[0][1])

def process(tempa, tempb, v, l):
    a1 = check_type(tempa)
    a2 = check_type(tempb)
    if a1 == "input error" or a2 == "input error":
        return "input error"

    if a1 == "function" or a2 == "function":
        if a1 == "constant" or a2 == "constant":
            return "no"
        else:
            if a1 == "variable" and a2 == "function":
                if tempa in tempb:
                    return "no"
                else:
                    l.append([tempa, tempb])
                    if v != []:
                        subst(v, l)
                    v.append([tempa, tempb])
                    return "top"
            elif a2 == "variable" and a1 == "function":
                if tempb in tempa:
                    return "no"
                else:
                    l.append([tempb, tempa])
                    if v != []:
                        subst(v, l)
                    v.append([tempb, tempa])
                    return "top"
            else:
                f1 = check_funname(tempa)
                f2 = check_funname(tempb)
                t1 = check_numarg(tempa)
                t2 = check_numarg(tempb)

                if t1 == "input error" or t2 == "input error":
                    return "input error"

                if f1 == f2 and t1 == t2:
                    count = 0
                    arg1 = ret_arg(tempa)
                    arg2 = ret_arg(tempb)
                    while count < t1:
                        var = process(arg1[count], arg2[count], v, l)
                        if var == "top":
                            return "top"
                        elif var == "no":
                            return "no"
                        elif var == "input error":
                            return "input error"
                        elif var == "True" and count == (t1 - 1):
                            count += 1
                            return "True"
                        else:
                            count += 1
                else:
                    return "no"

    elif a1 == "constant" and a2 == "constant":
        if tempa == tempb:
            return "True"
        else:
            return "no"

    elif a1 == "variable" or a2 == "variable":
        if a1 == "variable":
            if a2 == "variable":
                if tempa == tempb:
                    return "True"
                else:
                    l.append([tempa, tempb])
                    if v != []:
                        subst(v, l)
                    v.append([tempa, tempb])
                    return "top"
            else:
                l.append([tempa, tempb])
                if v != []:
                    subst(v, l)
                v.append([tempa, tempb])
                return "top"
        else:
            l.append([tempb, tempa])
            if v != []:
                subst(v, l)
            v.append([tempb, tempa])
            return "top"

def main():
    a = input("Enter first argument: ")
    b = input("Enter second argument: ")
    v = []
    l = []
    var = "top"

    while var == "top":
        if len(l) == 0:
            var = process(a, b, v, l)
        else:
            a = a.replace(l[0][0], l[0][1])
            b = b.replace(l[0][0], l[0][1])
            l = []
            var = process(a, b, v, l)

    if var == "no":
        print("no")
        v = []
    elif var == "input error":
        print("input error")
        v = []
    elif var == "True":
        print("Yes")
        for i in range(len(v)):
            print(f"{v[i][0]} --> {v[i][1]}")

if __name__ == "__main__":
    main()
```
